# 4-3 통계적 추론

```
import numpy as np
from scipy import stats

x_bar, mu, sigma, n = 196, 200, 5.3, 20

t_sample = (x_bar - mu) / (sigma / np.sqrt(float(n)))
print('검정 통계량 : ', np.round(t_sample, 2))

alpha = 0.05
t_alpha = stats.t.ppf(alpha, n-1)
print('t-table로부터의 임계값 : ', np.round(t_alpha, 3))

p_val = stats.t.sf(np.abs(t_sample), n-1)
print('t-table의 아래쪽 꼬리 p값 : ', np.round(p_val, 4))
```

```
import numpy as np
from scipy import stats

x, mu, sigma = 75, 56, 13.6

z = (x - mu) / sigma
print('z-score : ', np.round(z, 2))

p_val = 1 - stats.norm.cdf(z)
print('학생이 {}점 이상 받을 확률 : {}'.format(x, np.round(p_val*100, 2)))
```

```
import numpy as np
import pandas as pd
from scipy import stats

np.random.seed(0)

smoke = ['안함', '가끔', '매일', '심함']
exercise = ['안함', '가끔', '매일']

data = {'smoke' : np.random.choice(smoke, size=500),
        'exercise' : np.random.choice(exercise, size=500)}
df = pd.DataFrame(data, columns=['smoke', 'exercise'])
df.head()
```

```
contg = stats.chi2_contingency(observed=xtab)
p_val = np.round(contg[1], 3)
print('P-값 : ', p_val)
```

# 5-4 하이퍼 파라미터 선택

*   파라미터 - 학습하는 과정에서 자동으로 설정되는 매개변수
*   하이퍼 파라미터 - 개발자가 모델을 제어하기 위해서 사용하는 값



```
train_accuracy = []
test_accuracy = []

del range # 'numpy.int64' object is not callable // 이미 (예를들어) sum을 변수명으로 사용한 후에 sum() 함수를 다시 사용할 경우 이름의 중복으로 인해 해당 오류가 발생하게 된다

neighbors = range(1, 11)
for n in neighbors:
    knn = KNeighborsClassifier(n_neighbors = n)
    knn.fit(X_train, y_train)
    train_accuracy.append(knn.score(X_train, y_train))
    test_accuracy.append(knn.score(X_test, y_test))
```

```
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(neighbors, train_accuracy, label='train accuracy')
plt.plot(neighbors, test_accuracy, label='test accuracy')
plt.xlabel('n_neighbors')
plt.ylabel('accuracy')
plt.legend();
```

# 6-1 머신러닝 교차검증 개념 및 기법

교차검증의 기본 절차
1. 교차검증 1단계에서 데이터를 학습용과 테스트용으로 나눔
2. 모델의 테스트 성능을 기록
3. 교차 검증의 매 단계마다 다른 파티션으로 위의 작업을 수행
4. 모델의 최종 성능은 매 단계의 테스트 성능을 평균으로 계산

교차검증은 모델의 변동성을 줄여주며 모버피팅과 같은 문제를 막아줌

```
# K폴드(겹) 교차검증 
''' 
데이터를 무작위로 k개의 동일한 크기인 폴드로 나누고 
각 과정에서 특정 폴드를 테스트 용으로 나머지를 학습용으로 사용
이 과정을 k번 반복 시행
최종적으로 모델 성능의 평균을 계산
'''

from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target

from sklearn.neighbors import KNeighborsClassifier
model = KNeighborsClassifier(n_neighbors=1)

from sklearn.model_selection import cross_val_score 
cross_val_score(model, X, y, cv = 5) # 다섯번 반복, 다섯개의 폴드
```

```
# 단일 관측치 제거 방식 (LOOCV)
'''
Leave-one-out cross validation
매 시행 단계에서 테스트 샘플을 고정하는 방식.
검증을 시행할 때마다 한 지점을 제외한 모든 지점에서 훈련. 

데이터를 n개의 서브세트로 분할하고 
n개 중 1개를 테스트용으로 두고 n-1개로 학습을 수행
데이터 크키가 n이면 n번의 교차 검증을 수행 (따라서 데이터의 크기가 중요함!)
'''

from sklearn.model_selection import LeaveOneOut
scores = cross_val_score(model, X, y, cv = LeaveOneOut())
scores
```

# 6-3 편향-분산 트레이드 오프

1. 언더 피드 - 과소 적합, 모델이 고편향 됨. 모델이 모든 특징을 적정히 설명할 수 있을만큼의 유연성이 충분하지 않음.
2. 오버 피트 - 과대 적합, 모델이 고분산 됨. 유연성이 충분하지만 훈련 데이터의 잡음까지 반영하고있음.

```
from sklearn.preprocessing import PolynomialFeatures # 차수 증가
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import make_pipeline # 파이프라인에 등록하면 객체들이 seamless하게 수행되어 생산성이 높아짐
import numpy as np

np.random.seed(1)
X = np.random.rand(400, 1) ** 2 # 400행 1열 // 제곱
y = (10 - 1. / (X.ravel() + 0.1)) + np.random.rand(400) # + 오차 추가
```

```
%matplotlib inline
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')

X_test = np.linspace(-0.1, 1.1, 500).reshape(-1, 1) # 2차원 배열로 변경

fig = plt.figure(figsize=(16, 10))
for i, degree in enumerate([1, 3, 5, 10], start=1): # degree가 enumerate정보라서 i 인덱스 생성 // 인덱스를 1로 시작시킴
    ax = fig.add_subplot(2, 2, i)
    ax.scatter(X.ravel(), y, s = 5) # revel을 이용하여 2차원 X를 1차원으로 플랫시킴
    y_test = make_pipeline(PolynomialFeatures(degree), LinearRegression())\
    .fit(X, y).predict(X_test) # 파이프라인 앞단에서 피처 엔지니어링 작업 (차수 조정) # 테스트 예측값으로 라인 그래프 그리기
    ax.plot(X_test.ravel(), y_test, label='degree={0}'.format(degree))
    ax.set_xlim(-0.1, 1.2)
    ax.set_ylim(-1, 12)
    ax.legend(loc='best');
```

검증 곡선
- 편향과 분산 사이의 트레이드 오프에서 가장 효율적인 지점 -> 최적 모델

```
plt.figure(figsize=(10, 6))
plt.scatter(X.ravel(), y)
lim = plt.axis()
y_test = make_pipeline(PolynomialFeatures(3), \
                       LinearRegression()).fit(X, y).predict(X_test)
plt.plot(X_test.ravel(),y_test);
plt.axis(lim);
```

# 7-5 단순선형회귀

*   연속형(수치형) 반응 변수의 값을 예측하는 것
*   설명 변수가 하나(단일 특징) 단일 반응 변수 간에 선형관계가 있다고 가정하고 초평면이라고하는 선형 평면을 의용해 모델링한 것
*   단순회귀는 설명 변수의 1차원 반응변수의 1차원, 총 2개 차원을 가지며, 초명면은 한차원 낮은 1차원이 됨 (1차원 초평면은 선)
*   y = a + bx // y = 반응변수의 예측값 // a, b = 알고리즘을 통해 학습하게되는 파라미터 // x = 설명 변수, 특징


최소 제곱법 
1. OLS(Ordinary Least Squares) 또는 LLS(Linear Least Squares)로 부름
2. 단순 선형 회귀에서 모델을 최적화하는 파라미터 값을 훈련데이터로부터 학습하는 방법
3. 비용함수 : (손실함수) 모델의 오차를 정의하고 측정하기 위해 사용
4. 잔차 : (residual) 훈련 오차로 훈련 데이터의 관측 값과 모델에 의한 예측 값의 차이
5. 예측 오차 : (prediction error) 테스트 오차로 테스트 데이터의 관측 값과 모델에 의한 예측 값의 차이
6. 잔차의 합을 최소화 -> 값을 예측할 수 있는 추정기를 만들 수 있음.

비용 함수를 사용한 모델의 적합화 척도 : RSS
1. 반응 변수의 예측값이 훈련 데이터의 관측 값과 가까워지면 모델이 적합화되었다고 할 수 있다
2. 모델 적합화 척도 : 잔차 제곱 합 (RSS - Residual Sum of Squares)

*   설명변수의 n차원 반응 변수의 1차원, 총 n+1차원을 가지며, 초평면은 한 차원 낮은 n차원이 됨
*   y = a + b1x1 + b2x2 + b3x3 + + bnxn



# 7-7 다항 선형 회귀

*   특징을 변환한 후 다중 선형 회귀와 같은 방식으로 설명 변수롸 반응 변수의 비선형(곡선) 관계를 모델링 함

* y = a + b1x +  b2x^2 + b3x^3 + + bnx^n

5. 오버피팅된 모델은 지나친 노이즈를 반영할 수 있으므로 모델을 단순화하여 좀 더 일반화된 모델로 만들 필요가 있음


# 8-5 리지 회귀 모델 개념 및 와인품질예측모델만들기

* Ridge 회귀 (Tikhonov 규제)는 규제가 추가된 선형회귀 모델
* 규제항이 비용함수에 추가됨
* 가중치 벡터의 제곱항을 합한 L2 노름(정규화)을 2로 나눈 값을 적용함
* 모델의 가중치가 가능한 작게 유지되도록 함
* 규제항은 훈련하는 동안에만 비용함수에 추가됨
* 모델의 훈련이 끝나면 모델의 성능을 규제가 없는 성능 지표로 평가함
* 하이퍼 파라미터 alpha는 모델 규제를 조절가
* alpha가 0이면 선형회귀와 동일하며, alpha가 아주 크면 모든 가중치가 0에 가까워짐

```
from sklearn.linear_model import  Ridge

model_name = 'ridge'
alpha = 0.05
fig = plt.figure(figsize=(10, 4))
ax = fig.add_subplot(111)
ridge = Ridge(alpha = alpha)
ridge.fit(X_train, y_train)
y_pred = ridge.predict(X_test)
rmse = np.round(np.sqrt(mean_squared_error(y_test, y_pred)), 3)
coef = pd.Series(data = ridge.coef_, index = X_train.columns).sort_values()
ax.bar(coef.index, coef.values)
ax.set_xticklabels(coef.index, rotation = 45) # 변수명을 45도 돌림
ax.set_title('{0}: alpha = {1}, rmse = {2}'.format(model_name, alpha, rmse));
```

# 9-2 의사결정 트리의 특징

* 데이터를 분할하기 위해 가장 중요한 특징이 무엇인지 선택하는 로컬 최적화 방법을 계속 적용하는 탐욕적 방식으로 트리를 생성
* 의사결정 트리는 학습 샘플을 셋으로 분할하면서 생성되며, 분할의 과정은 각 서브 셋에 대해 재귀 형태로 진행됨
* 각 노드에서의 분할은 특징값을 기반으로 조건 검사를 통해 진행되며, 서브셋이 동일한 클레스 레이블을 가지는 경우나 분할을 통한 클래스 분류가 더 이상 의미 없을 경우 분할 작업을 마침침

# 9-3 의사결정트리 구현을 위한 CART 알고리즘

CART 알고리즘의 특징

* Classification and Regression Tree
* 노드를 왼쪽, 오른쪽 자식 노드로 부할 확장하면서 트리를 생성함
* 분할 단계에서 가장 중요한 특징과 해당 값의 모든 가능한 조합을 측정 함수를 이용해 탐욕적으로 찾음
* 범주형 특징의 경우 해당 특징값을가진 샘플들을 오른쪽 자식 노드에 할당함
* 수치형 특징의 경우 해당값보다 큰 값을 가진 샘플들을 오른쪽 자식 노드에 할당함

트리 분할의 측정 기준
* 한 노드에  모든 샘플이 같은 클래스에 있을 경우 순수하다고 함 ( gini == 0)

비용함수
* CART가 분류에 대해 최소화해야하는 비용함수
* 훈련세트의 분할은 최대 깊이가 되면 중지하거나, 샘플의 개수가 분할에 필요한 최소개수 이하일 경우나, 불순도를 줄이는 분할을 더이상 찾지 못 할 떄 중지 됨

# 9-7 의사결정트리 기반 광고클릭 예측 모델의 성능 측정

혼동행렬
* 실제 관측값을 얼마나 정확하게 예측했는지를 보여주는 행렬
* True Positive : 참-긍정은 병에 관해 '예'(병이 있을 것이다)로 예측한 환자가 실제로 병이 있는 경우
* True Negative : 참-부정은 병에 관해 '아니오'(병이 없을 것이다)로 예측한 환자가 실제로 병이 없는 경우
* Fales Positive :거짓-긍정은 병에 관해 '예'(병이 있을 것이다)로 예측한 환자가 실제로 병이 없는 경우 (1형 오류)
* Fales Negative :거짓-부정은 병에 관해 '아니오'(병이 없을 것이다)로 예측한 환자가 실제로 병이 있는 경우 (2형 오류)

분류 모델의 평가지표
* 정확도 : accuracy. 전체 샘플에서 전확하게 예측한 샘플 수의 비율
* 정밀도 : precision. Positive 클래스로 예측한 샘플에서 실제 Positive 클래스에 속하는 세플 수의 비율
* 재현율 : recall. 실제 positive 클래스에 속한 샘플에서 Positive 클래스에 속한다고 예측한 샘플 수의 비율 (참 긍정률 : True Positive rate. 민감도 : Sensitivity)
* 위양성율 : Fallout. 실제 positive 클래스에 속하지 않는 샘플에서 Positive 클래스에 속한다고 예측한 샘플 수의 비율 (거짓 긍정률)

* F1 점수 : f1 score. 정밀도와 재현율의 조화 평균. 점수에 상수 2를 곱해 정밀도와 재현율이 모두 1일 경우 점수가 1이 되도록 만듦
* 특이성 : specificity. 실제 Negative 클래스에 속한 샘플에서 Negative 클래스에 속한다고 예측한 샘플 수의 비율 (1 - Fales Positive Rate)

* AUC : Area Under the Curve. 곡선하면적
* 참 긍정률(TPR)과 거짓 부정률(FPR) 사이를 표현하기 위해 ROC(Receiver Operrating Characteristics)(수신자 조작 특성 곡선) 커브를 사용
* 예측된 확률로부터 여러 클래스로 분류를 수행하는데 활용

# 10-1 로지스틱 회귀 개념

로지스틱 회귀의 개요

* 베르누이 분포 : 긍정 확률 P와 부정 확률 1-P를 갖는 랜덤 변수의 확률 분포 (예를 들어 동전 던지기), (자연계의 일반적인 정규분포와 대칭이 되는 구조의 분포)
* 베르누이 분포를 따르는 반응 변수의 확률 범위는 [0, 1]
* 선형회귀는 특징값의 일정한 변화로 반응변수에도 일정한 변화가 일어난다는 가정을 하므로, 반응 변수가 확률을 나타낼 경우 가정이 유효하지 않음.
* 일반화 선형 모델은 연결 함수를 사용해 특징의 선형 조합을 반응 변수와 연결 짓는 방법을 통해 선형 회귀의 가정을 없앰
* 특징의 선형 조합을 정규분포를 따르지 않는 반응 변수와 연계하려면 연결함수가 필요

# 10-2 로지스틱 회귀의 특징

로지스틱 회귀와 분류
* 로지스틱 회귀의 반응 변수의 값 = 긍정(양성) 클래스의 확률
* 반응 변수의 값 >= 임계치 -> 긍정(양성) 클래스를 예측
* 반응 변수의 값 <= 임계치 -> 부정(음성) 클래스를 예측
* 반응 변수는 '로지스틱함수'를 사용해 특징의 선형 조합 합수로 모델링 됨
* e는 오일러 상수(2.718....무리수), t는 성명 변수의 조합(B0 + Bx)

로지스틱 함수 = 시그모이드 함수

```
def sigmoid(z):
    return 1.0 / (1 + np.exp(-z)) 
```

# 12 나이브 베이즈

# 12-1 나이브 베이즈 개념 및 스팸매일 필터

나이브 베이즈 분류기란?
* 확률 기반 분석기
* 데이터가 각 틀래스에 속할 특징 확률을 계산
* 나이브(Naive) : 예측한 특징이 상호 독립적이라는 가정 하에 확률 계산을 단순화함
* 베이즈(Bayes) : 입력 특징이 클래스 전체의 확률 분포 대비 특정 클래스에 속할 확률을 베이즈 정리를 기반으로 계산

베이즈 정리와 나이브 베이즈 동작 원리 이해
* 베이즈 정리에서 P(A|B)는 사건 B가 주어졌을 때 사건 A가 일어날 확률
* 나이브 베이즈의 목표는 n개의 특징을 가진 샘플데이터 x가 주어졌을 때 k개의 클래스 중 하나에 속할 확률을 결정하는 것
* 샘플 데이터는 x1, x2, x3, ..., xn의 값을 가진 특징으로 구성되고 , yk는 샘플 데이터가 k에 속하는 사건을 나타냄 -> 베이즈 정리를 이용

* 샘플 데이터가 k에 속하는 사건의 확률 P(yk)는 관측값의 특징에 대한 지식 없이 클래스가 어떻게 분포되어 있는지 나타냄 -> '사전 확률'
* 사전 확률은 사전에 결정되어있거나, 핛ㅂ 샘플 데이터를 이용해 학습시킬 수도 있음
* P(yk|x)는 관측값에 대해 외부 지식을 이용한 '사후 확률'
* P(x|yk)는 클래스 yk에 속한 샘플 데이터가 주어졌을 때 n개의 특징에 대한 결합 분포를 나타냄 -> 유사 가능도

* 나이브 베이즈는 특징 간 서로 독립이라는 가정이 있으므로, n개의 특징에 대한 결합 조건부 분포는 특징의 분포를 곱한 것으로 표현할 수 있음
** P(x|yk) = P(x1|yk) * P(x2|yk) * P(x3|yk) * ... * P(xn|yk)
* P(x)는 특정 클래스에 속하지 않은 특징의 분포에 따라 값이 계싼되므로 상수로 처리할 수 있음
* 사전 확률과 유사 가능도를 이용해 사후 확률을 계산할 수 있음
**  
* 라플라스 스무딩 : Laplace smoothing. 특징의 출현 횟수 초기값을 1부터 시작해 0을 곱해 발생하는 문제를 해결 (발견되지 않은 특징의 출현 빈도 초기값을 1로 설정)

```
import glob
import os
import numpy as np

emails, labels = [], []

paths = ['/content/drive/MyDrive/ColabCSV/enron1/spam/', '/content/drive/MyDrive/ColabCSV/enron1/ham/']
for path in paths:
    for filename in glob.glob(os.path.join(path, '*.txt')):
        with open(filename, 'r', encoding = 'ISO-8859-1') as file:
            emails.append(file.read())
            if path.endswith('spam/'):
                labels.append(1)
            else:
                labels.append(0)

print(np.unique(labels, return_counts=True))
```

# 13 추천 엔진

# 13-1 추천 엔진의 개념과 종류류

추천 엔진의 개요
* 제품 정보와 사용자 정보와 같은 엄청난 양의 데이터를 분석하고 데이터 마이닝 방식을 기반으로 연관 추천을 제공하는 기술이자 도구
* 사용자가 얼마나 해당 항목을 좋아할지 예측할 수 있는 수학적 모델 또는 목적함수를 개발하는 것이 중요
* 많은 선택 가능한 제품들 중 가장 좋은 결정을 내릴 수 있도록 적절한 추천을 하는 것이 목표

추천 엔진의 종류
* 협업 필터링 추천 시스템 : 사용자 선호도를 이용해서 선택 가능한 많은 집합들로부터 아이템을 필터링함
 - 사용자 기반 협업 필터링 : 사용자 주변의 취향을 고려해 추천
    1. 유사한 선호도를 가진 사용자를 발견
    2. 활성 사용자는 아직 등급을 매기지 않은 아이템에 대해 유사한 사용자가 제공한 등급을 참고해 활성 사용자에게 새로운 아이템을 추천
 - 아이템 기반 협업 필터링 : 아이템의 유사 제품을 이용해 추천
    1. 아이템 선호도를 기반으로 아이템 유사도 계산
    2. 활성 사용자가 과거 평가한 아이템과 가장 유사한 아직 평가하지 않은 아이템을 찾아서 아이템 추천
- 콘텐츠 기반 추천 시스템 : 아이템의 콘텐츠 정보를 활용. 아이템 속성과 해당 아이템 속성에 대한 사용자 선호도를 사용해 추천
- 하이브리드 추천 시스템 : 콘텐츠 기반 추천 시스템 + 협업 필터링 추천 시스템
- 상황 인식 추천 시스템 : 추천 항목을 계산하기 전에 시각, 계절, 분위기, 장소 등의 상황을 고려하는 유형의 추천 시스템

# 13-2 추천 엔진 개발을 위한 협업 필터링 개념

협업 필터링 개요
- 어떤 특정인이 새로운 아이템에 대해 어떤 선호도를 갖는지 예측하고자할 때, 그 아이템에 대한 다른 다수의 선호도 집합을 활용하는 방법

- 아이템의 내용을 반드시 알아야 할 필요가 없음

```
from sklearn.metrics import mean_squared_error

def get_mse(pred, actual):
    pred = pred[actual.nonzero()].flatten()
    actual = actual[actual.nonzero()].flatten()
    return mean_squared_error(pred, actual)
```
